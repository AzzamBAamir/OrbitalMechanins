```python
#Evaluating the path of the Earth around the sun

import matplotlib.pyplot as plt
from numpy import arange,exp
from math import cos, sin, pi, sqrt

#Defining Important Variables
GM_S = 4 * (pi**2)
a = 0 #Limits of plot in AU
b = 20 
h = 0.002
vx = 0.00
vy = 2 * pi

#Initial Starting point for Earth given Sun (0,0)
x = 1.0
y = 0.0

xvals = []
yvals = []
rvals = []

for i in arange (a,b,h):  #Euler-Cromer Method
    r = ((x**2)+(y**2))**(1/2)
    vx -= ((4*(pi**2))*x)/(r**3) * h
    vy -= ((4*(pi**2))*y)/(r**3) * h
    x += vx * h
    y += vy * h
    rvals.append(r)
    xvals.append(x)
    yvals.append(y)
plt.plot(xvals,yvals)
plt.xlabel("x(AU)")
plt.ylabel("y(AU)")
plt.title("Earth's orbit plotted using Euler-Cromer Method")
plt.show()

```

```python
#Evaluating the precession of Mercury

import matplotlib.pyplot as plt
from numpy import arange,exp, diff
from math import cos, sin, pi, sqrt, acos, isclose
import scipy.optimize as optimization
from scipy.stats import linregress

#Use method of least squares to fit line to our data
#Defining Important Variables
GM_S = 4 * (pi**2)
a = 0 
b = 2
h = 0.00001
vx = 0.00
vy = 8.2
alpha = 0.0008
t = 0.0

def fit(x,a,b):
    return (a*x) + b 

#Initial Starting point for Mercury given Sun (0,0)
x = 0.47
y = 0.0

alpharange = arange(0,0.008,0.001)
alphavals = []
gradientvals1 = []

for i in alpharange:
    alphavals.append(i)
    xvals = []
    yvals = []
    rvals = []
    angles = []
    time = []
    #Get r at every loop? that could help
    for i in arange (a,b,h):  #Euler-Cromer Method
        r = ((x**2)+(y**2))**(1/2)
        if r >= 0.47:
            theta = acos(x/r) * (180/pi)
            angles.append(theta)
            time.append(i)
        vx -= (((4*(pi**2))*x)/(r**3))*(1+(alpha/(r**2))) * h
        vy -= (((4*(pi**2))*y)/(r**3))*(1+(alpha/(r**2))) * h
        x += vx * h
        y += vy * h
        xvals.append(x)
        yvals.append(y)
    #print(linregress(time,angles))
    popt,pcov= (optimization.curve_fit(fit,time,angles))
    gradientvals1.append(popt[1])
plt.plot(alphavals,gradientvals1,label="CurveFit")
plt.xlabel("alpha")
plt.ylabel("dtheta/dt (degrees/year)")
plt.title("Simulation of the precession of Mercury")
plt.legend(loc='best')
plt.show()
print(linregress(alphavals,gradientvals1))
```

```python
from math import exp
from numpy import arange, log
import matplotlib.pyplot as plt
import scipy.optimize as optimization
from scipy.stats import linregress

#First plot exact solution

N_0 = 100
tau = 1
h = 0.005
a = 0.0
b = 5.0

def fit(x,a,b):
    return (a*x) + b

def decay(t):
    return N_0*(exp(-t/tau))

time = []
yval = []
for i in arange(a,b,h):
    yval.append(decay(i))
    time.append(i)
plt.xlabel("Time")
plt.ylabel("Number of Nuclei")
plt.title("Number of Nuclei vs Time")
plt.plot(time,yval)

#Now doing Runge-Kutta 4 method

def derivative(Nuclei,t):
    return Nuclei/tau
yval1 = []
time = []
N = 100
for i in arange(a,b,h):
    time.append(i)
    yval1.append(N)
    k1 = h*derivative(N,i)
    k2 = h*derivative(N+(0.5*k1),i+(0.5*h))
    k3 = h*derivative(N+(0.5*k2),i+(0.5*h))
    k4 = h*derivative(N+k3,i+h)
    N -= (k1+(2*k2)+(2*k3)+k4)/6
plt.plot(time,yval1,'k')
plt.show()

timesteps = arange(0.01,1,0.01)
error = []
for h in timesteps:
    time = []
    yval1 = []
    yval = []
    N = 100
    t = []
    for i in arange(a,b,h):
        time.append(i)
        yval.append(decay(i))
        yval1.append(N)
        k1 = h*derivative(N,i)
        k2 = h*derivative(N+(0.5*k1),i+(0.5*h))
        k3 = h*derivative(N+(0.5*k2),i+(0.5*h))
        k4 = h*derivative(N+k3,i+h)
        N -= (k1+(2*k2)+(2*k3)+k4)/6
    for i in range(len(yval)):
        t.append(abs(yval1[i]-yval[i]))
    error.append(max(t))
plt.plot(timesteps,error)
plt.xlabel("timesteps, dt")
plt.ylabel("error")
plt.title("Max error plotted against timestep")
plt.show()
print(linregress(timesteps,error))
plt.plot(timesteps,log(error))
plt.show()
print(linregress(timesteps,log(error)))
```
